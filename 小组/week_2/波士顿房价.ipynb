```python
x_data = datas[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']] # 导入所有特征变量
y_data = datas[['PRICE']] # 导入目标值（房价）

plt.figure(figsize=(18,10))

for i in range(13):
    plt.subplot(4,4,i+1)
    plt.scatter(x_data.values[:,i],y_data,s = 5)    #.values将DataFrame对象X_df转成ndarray数组
    plt.xlabel(names[i])
    plt.ylabel('Price')
    plt.title(str(i+1)+'. '+names[i]+' - Price')  
    
plt.tight_layout()
plt.show()
```

#### 通过不同数据的点图判断其与房价的相关性，亦可得，房价数量是与房价最相关的数据

## 设置训练集和测试集

```python
x_train, x_test, y_train, y_test = train_test_split(x_data,
                                                    y_data,
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
ones = np.ones((1,404))
X = x_train.values
X = np.concatenate((ones, X.T))
X = X.T
Y = y_train
```

```python
print(Y)
Y.shape
```

```python
print(X)
X.shape
```

```python
theta = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), Y)
```

```python
print(theta)
theta.shape
```

## 根据预测的theta值查看测试集的误差

```python
ones = np.ones((1,102))
X = x_test.values
X = np.concatenate((ones, X.T))
X = X.T
Y = y_test
Y_hat = np.dot(X, theta)
```

```python
plt.figure(figsize=(30, 10))  # 设置画板尺寸
plt.title('波士顿房价预测曲线与实际曲线对比图', fontsize=15)
x = range(len(Y))  # x轴数据

plt.plot(x, np.array(Y.T)[0], color='r', label='实际价格')  # 实际价格曲线
plt.plot(x, np.array(Y_hat.T)[0], color='g', ls='--', label='预测价格')  # 预测价格曲线
plt.legend(fontsize=30, loc=1)  # 显示图例
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.title('波士顿房价拟合直线', fontsize=20)
plt.plot(range(40),range(40),ls='--')
plt.scatter(np.array(Y.T)[0], np.array(Y_hat.T)[0])
plt.xlabel('实际价格')
plt.ylabel('预测价格')
plt.show()
```
